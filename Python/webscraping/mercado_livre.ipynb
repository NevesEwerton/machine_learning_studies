# Mercado Livre

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
urlbase = 'https://lista.mercadolivre.com.br/'

produto_nome = input("Qual produto você deseja? ")

response = requests.get(urlbase + produto_nome)

site = BeautifulSoup(response.text, 'html.parser')

produtos = site.findAll('div',attrs={'class':'andes-card ui-search-result ui-search-result--core andes-card--flat andes-card--padding-16'})

for produto in produtos:

    titulo = produto.find('h2',attrs={'class':'ui-search-item__title'})

    link = produto.find('a', attrs={'class':'ui-search-link'})

    real = produto.find('span',attrs={'class':'andes-money-amount__fraction'})

    centavos = produto.find('span',attrs={'class':'andes-money-amount__cents andes-money-amount__cents--superscript-24'})

    #print(produto.prettify())
    print("Titulo do produto: ",titulo.text)
    print("Link do produto: ", link['href'])
    
    if centavos:
        print("Preço do produto: R$",real.text + ',' + centavos.text)
    else:
        print("Preço do produto: R$",real.text)

    print('\n\n')
```

```python
urlbase = 'https://lista.mercadolivre.com.br/'

produto_nome = input("Qual produto você deseja? ")

response = requests.get(urlbase + produto_nome)

site = BeautifulSoup(response.text, 'html.parser')

produtos = site.findAll('div',attrs={'class':'andes-card ui-search-result ui-search-result--core andes-card--flat andes-card--padding-16'})

lista = []

for produto in produtos:

    titulo = produto.find('h2',attrs={'class':'ui-search-item__title'})

    link = produto.find('a', attrs={'class':'ui-search-link'})

    real = produto.find('span',attrs={'class':'andes-money-amount__fraction'})

    centavos = produto.find('span',attrs={'class':'andes-money-amount__cents andes-money-amount__cents--superscript-24'})
        
    if centavos:
        lista.append([titulo.text, real.text + ',' + centavos.text, link['href']])
    else:
        lista.append([titulo.text,real.text, link['href']])

data = pd.DataFrame(lista, columns=['Produto','Preço','Link'])
data.to_excel('lista_pulseira_galaxy_watch.xlsx', index=False)
```

```python
urlbase = 'https://www.amazon.com.br/s?k='

produto_nome = input("Qual produto você deseja? ")

response = requests.get(urlbase + produto_nome)

site = BeautifulSoup(response.text, 'html.parser')

produtos = site.findAll('div',attrs={'class':'a-section a-spacing-base'})

for produto in produtos:
    
    titulo = produto.find('span',attrs={'class':'a-size-base-plus a-color-base a-text-normal'})

    real = produto.find('span',attrs={'class':'a-price-whole'})

    centavos = produto.find('span',attrs={'class':'a-price-fraction'})

    estrelas = produto.find('span', attrs={'class':'a-icon-alt'})
    
    print(titulo.text)
    print(real.text + centavos.text)
    print(estrelas.text)
```

```python
lista = []

urlbase = 'https://www.amazon.com.br/s?k='

produto_nome = input("Qual produto você deseja? ")

response = requests.get(urlbase + produto_nome)

site = BeautifulSoup(response.text, 'html.parser')

produtos = site.find_all('div',attrs={'class':'a-section a-spacing-base'})

for produto in produtos:
    
    titulo = produto.find('span',attrs={'class':'a-size-base-plus a-color-base a-text-normal'})

    real = produto.find('span',attrs={'class':'a-price-whole'})

    centavos = produto.find('span',attrs={'class':'a-price-fraction'})

    estrelas = produto.find('span', attrs={'class':'a-icon-alt'})
    
    lista.append([titulo.text, real.text+centavos.text, estrelas.text, "Amazon"])
    

df = pd.DataFrame(lista, columns=['Produto','Preço','Nota','Empresa'])
df.to_excel('WP_amazon.xlsx', index=False)
```
